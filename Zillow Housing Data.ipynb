```python
import scipy as sci
import numpy as np
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
```

```python
zillow= "C:/Users/JOEBO/Desktop/Data Science projects/Zillow Home Data/ZHVI.csv"
house = pd.read_csv(zillow)
house.rename(columns={'Unnamed: 0' : 'Date'}, inplace = True)
```

```python
#What may have caused the fall of the market in 2008 following a sharp rise?
#Is the current housing market mimicking the past?
df1 = house[['Texas', 'Florida', 'New York']]
dt = house['Date']
d = pd.to_datetime(dt)
df1.set_index(d)
plt.plot(d,df1)
plt.legend(['Texas', 'Florida', 'New York'])
plt.gcf().autofmt_xdate()
plt.show()
```

```python
#It is presumed that lenient bank lending practices led to the crash in 2008
#Are there similarities between the practices before and after the market crash?
#Observing data from New York mortgage loans 2004-2016 will provide information on
#the amount of mortgages issued each year in New York by the State of NY Mortgage Agency
import seaborn as sns
import sklearn
nymortgage = pd.read_csv('C:/Users/JOEBO/Desktop/Data Science projects/New York Mortgage Data/state-of-new-york-mortgage-agency-sonyma-loans-purchased-beginning-2004-1.csv')
nymortgage.sort_values(by=['Purchase Year'], inplace = True)
nymortgage['Number of Units'] = nymortgage['Number of Units'].astype(str).str.replace(' Family', '')
nymortgage['Number of Units'] =nymortgage['Number of Units'].astype(float)
nymortgage['Number of Units']
loan_count = [nymortgage.loc[nymortgage['Purchase Year']== 2004, 'Purchase Year'].count(),
nymortgage.loc[nymortgage['Purchase Year']== 2005, 'Purchase Year'].count(),
nymortgage.loc[nymortgage['Purchase Year']== 2006, 'Purchase Year'].count(),
nymortgage.loc[nymortgage['Purchase Year']== 2007, 'Purchase Year'].count(),
nymortgage.loc[nymortgage['Purchase Year']== 2008, 'Purchase Year'].count(),
nymortgage.loc[nymortgage['Purchase Year']== 2009, 'Purchase Year'].count(),
nymortgage.loc[nymortgage['Purchase Year']== 2010, 'Purchase Year'].count(),
nymortgage.loc[nymortgage['Purchase Year']== 2011, 'Purchase Year'].count(),
nymortgage.loc[nymortgage['Purchase Year']== 2012, 'Purchase Year'].count(),
nymortgage.loc[nymortgage['Purchase Year']== 2013, 'Purchase Year'].count(),
nymortgage.loc[nymortgage['Purchase Year']== 2014, 'Purchase Year'].count(),
nymortgage.loc[nymortgage['Purchase Year']== 2015, 'Purchase Year'].count(),
nymortgage.loc[nymortgage['Purchase Year']== 2016, 'Purchase Year'].count()]

r=[2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016]
s = loan_count
plt.plot(r,s)
```
